```python
import numpy as np
import pandas as pd

#this page is extra stuff / setup
data = np.genfromtxt('ccp.csv', delimiter=',', dtype=np.uint, skip_header=1) 
#fancy indexing:
def mask_function(dataset, mask):
    data_indices = np.where(mask)
    res = dataset[data_indices]
    return res

def column(matrix, i, lookup_column, req):
    accum = 0
    for row in matrix:
        if row[lookup_column] == req:
            accum += row[i]
    return accum



#5180: German, 5130: Frankrig
german = 5180; french = 5130;
french_germs = mask_function(data, np.logical_or(data[:,3]==5180, data[:,3]==5130))
years_total = np.unique(french_germs[:,0])




def distribute_values(y_axis, column_req, accum_val_1, accum_val_2):
    accum_french = []
    accum_germ = []
    for yt in y_axis:
        germs = french_germs[(french_germs[:,column_req] == yt) & (french_germs[:,accum_val_1] == german)]
        french_number = column(french_germs, accum_val_2, column_req, yt)
    
        accum_french.append(french_number)    
        accum_germ.append(sum(germs[:,accum_val_2]))
    return accum_french, accum_germ
    
#unique years, year column, stat_code column for germans, amount for french
accum_french, accum_germ = distribute_values(years_total, 0, 3, 4)

#hardcoded
#for yt in years_total:
#    germs = french_germs[(french_germs[:,0] == yt) & (french_germs[:,3] == german)]
#    french_number = column(french_germs, 4, 0, yt)
    
#    accum_french.append(french_number)    
#    accum_germ.append(sum(germs[:,4]))


df = pd.DataFrame({'germans':accum_germ, 'french':accum_french}, index=years_total)
df.plot.bar()
```

```python
#Create a bar plot that shows distribution of german and british citizens in terms of age.
age_citiz = np.unique(french_germs[:,2])

accum_french, accum_germ = distribute_values(age_citiz, 2, 3, 4)

df = pd.DataFrame({'germans': accum_germ, 'french':accum_french}, index=age_citiz)
df.plot.bar(figsize=(40,50))
```

```python
#How many French are there compared to Germans and how are they distributed over various neighbourhoods?
neighborhood = np.unique(french_germs[:,1])

accum_french, accum_germ = distribute_values(neighborhood, 1, 3, 4)

df = pd.DataFrame({'germans': accum_germ, 'french':accum_french}, index=neighborhood)
df.plot.bar()
```
