```python
import requests
import re
import os
import pandas as pd
#still cannot import this from my prev notebook ... 
#https://gist.github.com/Thomas-Rosenkrans-Vestergaard/030b37343c2aa33e58aeb1ace95699dd
#Part 1
#Programatically download the raw csv data.

def download(url, to):
    response = requests.get(url)
    directory = ''.join(re.split('[/\\\]', to)[:-1])
    os.makedirs(directory, exist_ok=True)
    with open(to, "wb") as f:
        f.write(response.content)
        return to
download('https://raw.githubusercontent.com/PranayMalhotra/NBA-player-career-lengths/master/nba_logreg.csv', 'temp/nba_logreg.csv')

#Import the data into a pandas dataframe.
data = pd.read_csv(r'temp/nba_logreg.csv')
df = pd.DataFrame(data)
#Print the head of the dataset.
df.head()
```

```python
#Part 2
#Find the names of the top 10 3-point shooters (3p_made) in the dataset.
df_sorted = df.sort_values(['3p_made'], ascending=False)
df_sorted.head(10)
```

```python
#Part2
#On a scatter plot, show the relationship between x=GP and y=target_5yrs.
import matplotlib.pyplot as plt
plt.scatter(x=df['gp'].values, y=df['target_5yrs'].values) #what the luls?
#Does the scatter plot show a relationship between the two attributes?
#if you have an average of 35~ gp you're almost certain to pertain the target 5 years.
```

```python
#Part3
#Create a classifier that can predict whether or not a player will play for 5 years in the NBA.
#I chose: GP, MIN and PTS as the weighters 
from sklearn import tree
from sklearn.model_selection import train_test_split  

data_set = []
pertained_result = []
for x in range(len(df['gp'])):
    data_set.append([df['gp'][x], df['min'][x], df['pts'][x]])
    
    pertained_result.append(df['target_5yrs'][x])
x_train, x_test, y_train, y_test = train_test_split(data_set, pertained_result)

clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train, y_train)
pred = clf.predict(x_test)
print(pred)
```

```python
#Use your classifier, find examples where the result is both 0 and 1.
import matplotlib.pyplot as plt
#reading from prev, I can see read the respective values
print(clf.predict([x_test[0]])) #result is 0
print(clf.predict([x_test[2]])) #result is 1

#Reflect on the importance of the GP attribute. Is it more or less important, than the other attributes.
#it shows a players who fall below 35~ games on average has a high statistical change of getting replaced. Making it a requirement to be above.
```
